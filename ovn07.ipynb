## Recap - Object-Oriented Programming

```python
from abc import ABC, abstractmethod
import math

# ABSTRACT CLASS - CANNOT INSTANCIATE OBJECTS OF THIS CLASS
class Polygon(ABC):
  # CONSTRUCTOR
  def __init__(self, id):
    self.id = id

  def __str__(self):
    return "POLYGON "+ str(self.id) + " with area " + str(self.area())

  def get_id(self):
    return self.id

  # ABSTRACT METHODS - NEEDS TO BE OVERWRITTEN
  @abstractmethod
  def area(self):
    pass


class Rectangle(Polygon):
  def __init__(self, id, width, height):
    # CALL CONSTRUCTOR OF SUPERCLASS
    super().__init__(id)
    self.w = width
    self.h = height

  def area(self):
    return self.w * self.h

class Circle(Polygon):
  def __init__(self, id, radius):
    super().__init__(id)
    self.r = radius

  def area(self):
    return 2*math.pi*self.r*self.r

class Square(Rectangle):
  def __init__(self, id, side):
    super().__init__(id, side, side)
  

if __name__ == "__main__":
  rect = Rectangle(id=5, width=10, height=7)
  print(rect)

  circ = Circle(id=6, radius=5)
  print(circ)

  squ = Square(id=4, side=10)
  print(squ)
```
